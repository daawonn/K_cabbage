```python
import tensorflow as tf
import numpy as np
from pandas.io.parsers import read_csv
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = read_csv('price_data.csv', sep=',')
```

```python
data['year_1'] = data['year'].apply(lambda x: str(x)[:4])
data['month'] = data['year'].apply(lambda x: str(x)[4:6])
data.head()
```

```python
data = data[['year','avgTemp','minTemp','maxTemp','rainFall','year_1','month','avgPrice']]
data.head()
```

```python
minmax = MinMaxScaler()
```

```python
xy = np.array(data, dtype=np.float32)
yx = np.array(data.iloc[:, [-1]], dtype=np.float32 )
```

```python
xy = minmax.fit_transform(xy)
```

```python
model = tf.global_variables_initializer();
```

```python
# 4개의 변인을 입력을 받습니다.
x_data = xy[:, 1:-1]
```

```python
# 가격 값을 입력 받습니다.
y_data = yx
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data,y_data,random_state=0)
```

```python
# 플레이스 홀더를 설정합니다.
X = tf.placeholder(tf.float32, shape=[None, 6])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
# y = wX+b
W = tf.Variable(tf.random_normal([6, 1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")
```

```python
# 가설을 설정합니다. y = wX+b
# tensorflow 행렬곱 합수 matmul
hypothesis = tf.matmul(X, W) + b
# hypothesis = tf.nn.softmax(tf.matmul(X, W) + b)
```

```python
# 비용 함수를 설정합니다. (예측값 - 실제값)제곱의 평균
# reduce_mean은 특정 차원을 제거하고 평균을 구한다.
cost = tf.reduce_mean(tf.square(hypothesis - Y))
correct_prediction = tf.equal(tf.argmax(hypothesis,1),tf.argmax(Y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```python
# 최적화 함수를 설정합니다.
optimizer = tf.train.MomentumOptimizer(learning_rate=0.0005,momentum=0.9)
train = optimizer.minimize(cost)
```

```python
# 세션을 생성합니다.
sess = tf.Session()
```

```python
# 글로벌 변수를 초기화합니다.
sess.run(tf.global_variables_initializer())
```

```python
# 학습을 수행합니다.
for step in range(100001):
    cost_, hypo_, _ = sess.run([cost, hypothesis, train], feed_dict={X: X_train, Y: y_train})
    if step % 1000 == 0:
        print("#", step, " 손실 비용: ", cost_)
        print("- 배추 가격: ", hypo_[0])
        
# print("@", '정확도', accuracy.eval({X:x_data,Y:y_data}))        
```

```python
# 글로벌 변수를 초기화합니다.
sess.run(tf.global_variables_initializer())
```

```python
correct_prediction = tf.equal(tf.argmax(X_train,1),tf.argmax(y_train,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```python
for step in range(100001):
    cost_, hypo_,accuracy_, _ = sess.run([cost, hypothesis, accuracy, train], feed_dict={X: X_test, Y: y_test})
    if step % 1000 == 0:
        print("#", step, " 손실 비용: ", cost_)        
        print("- 배추 가격: ", hypo_[0])
        if step == 1000:
            print("@", '정확도', accuracy_)       
```

```python
# 학습된 모델을 저장합니다.
saver = tf.train.Saver()
save_path = saver.save(sess, "./saved.cpkt")
print('학습된 모델을 저장했습니다.')
```
